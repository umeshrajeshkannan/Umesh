```python
from ultralytics import YOLO
import cv2

# Load YOLOv8 model (pretrained)
model = YOLO("yolov8n.pt")   # nano version (fast)

# Open video (0 for webcam or use video file path)
cap = cv2.VideoCapture("/content/ambulance.png")

traffic_signal = "RED"

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)

    ambulance_detected = False

    for r in results:
        boxes = r.boxes
        for box in boxes:
            cls = int(box.cls[0])
            label = model.names[cls]

            if label == "truck":  # You must train custom model for ambulance
                ambulance_detected = True

    # Traffic Control Logic
    if ambulance_detected:
        traffic_signal = "GREEN"
    else:
        traffic_signal = "RED"

    # Display Signal Status
    cv2.putText(frame, f"Signal: {traffic_signal}", (50, 50),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    cv2.imshow("Smart Traffic System", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
